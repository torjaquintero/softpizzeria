## Ejercicio No. 8

1. try y except

    - Permite atrapar errores que ocurren durante la ejecución.

    - En este caso, cuando se espera un número entero, si el usuario ingresa algo distinto, se captura el error ValueError y se muestra un mensaje sin que el programa falle.

2. Validación de datos

    - Los campos de usuario y clave solo aceptan números enteros.

    - La opción de registrarse solo acepta 's' o 'n', evitando entradas inválidas.

3. Flujo del programa

    - Primero se intenta iniciar sesión con usuario y clave válidos.

    - Si el usuario no existe, se pregunta si desea registrarse.

    - Si se registra, se llama a la función registrarUsuario().

```python
# -----------------------------
# Variables iniciales para almacenar usuario y clave
# -----------------------------
usuarioGuardado = "admin"  # nombre de usuario permite letras
claveGuardada = 1234       # clave inicial de 4 dígitos

# -----------------------------
# Función para registrar un nuevo usuario
# -----------------------------
def registrarUsuario():
    # Nombre de usuario: se permite cualquier texto
    nuevoUsuario = input("Ingrese un nuevo usuario: ")
    
    # Clave: solo números enteros de 4 dígitos
    while True:
        nuevaClave = input("Ingrese una nueva clave (4 dígitos): ")
        try:
            nuevaClave = int(nuevaClave)  # verificar que sea un número
            if 1000 <= nuevaClave <= 9999:  # validar que tenga 4 dígitos
                break
            else:
                print("Error: la clave debe tener exactamente 4 dígitos.")
        except ValueError:
            print("Error: debe ingresar un número entero.")
    
    print(f"Usuario '{nuevoUsuario}' registrado correctamente.")
    return nuevoUsuario, nuevaClave  # devolver datos para actualizar variables

# -----------------------------
# Programa principal
# -----------------------------

# Solicitar nombre de usuario (puede ser texto)
usuarioIngresado = input("Ingrese su usuario: ")

# -----------------------------
# Validación de usuario
# -----------------------------
if usuarioIngresado == usuarioGuardado:
    intentos = 0            # contador de intentos de clave
    claveCorrecta = False   # bandera para indicar si la clave es correcta
    
    # Bucle para validar la clave, máximo 3 intentos
    while intentos < 3 and not claveCorrecta:
        try:
            claveIngresada = int(input("Ingrese su clave (4 dígitos): "))
            if 1000 <= claveIngresada <= 9999:  # validar que tenga 4 dígitos
                if claveIngresada == claveGuardada:
                    print("¡Acceso concedido! Bienvenido,", usuarioIngresado)
                    claveCorrecta = True
                else:
                    intentos += 1
                    print(f"Clave incorrecta. Intento {intentos} de 3.")
            else:
                print("Error: la clave debe tener exactamente 4 dígitos.")
        except ValueError:
            print("Error: debe ingresar un número entero.")
    
    # Si se agotaron los intentos de clave
    if not claveCorrecta:
        print("Ha excedido el número de intentos. Acceso denegado.")

# -----------------------------
# Usuario no registrado
# -----------------------------
else:
    while True:
        # Preguntar si desea registrarse
        respuesta = input("Usuario no encontrado. ¿Desea registrarse? (s/n): ").lower()
        if respuesta == 's':
            usuarioGuardado, claveGuardada = registrarUsuario()  # registrar usuario
            break
        elif respuesta == 'n':
            print("No se registró ningún usuario. Programa finalizado.")
            break
        else:
            # Validación de entrada: solo 's' o 'n'
            print("Error: solo puede ingresar 's' o 'n'.")
```
