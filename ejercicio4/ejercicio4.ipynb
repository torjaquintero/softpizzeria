1. Agrupación de datos: Creamos una tupla llamada menu que contiene todas las opciones. Una tupla es una colección ordenada e inmutable de elementos. Esto nos permite organizar los datos de manera más clara y facilita la manipulación.

2. Estructura repetitiva for:

    - for elementoTupla in menu: recorre cada elemento de la tupla menu.

    - Cada elemento se almacena temporalmente en la variable elementoTupla y se imprime con print().

    - Esto evita tener que escribir múltiples líneas print(opcion1), print(opcion2), etc., haciendo el código más compacto y fácil de mantener.

3. Interacción con el usuario: Después de imprimir todas las opciones, solicitamos la elección del usuario con input().

4. Condicionales: Se usan para mostrar qué opción ha sido elegida y validar que sea una de las opciones del menú.

```python
# Mensaje de bienvenida
Bienvenida = '\nBienvenidos a la pizzería de Sys On Chip\n'
print(Bienvenida)

# Título del menú
print('Menú principal\n')

# Opciones del menú
opcion1 = '1. Gestión de pedidos'
opcion2 = '2. Gestión de clientes'
opcion3 = '3. Gestión de empleados'
opcion4 = '4. Gestión de inventario'

# Agrupamos las opciones en una tupla
menu = (opcion1, opcion2, opcion3, opcion4)

# Imprimimos todas las opciones usando un ciclo for
for elementoTupla in menu:
    print(elementoTupla)

# Línea en blanco para separar la entrada del usuario
print('\n')

# Solicitar al usuario que elija una opción
opcionElegida = input('Qué opción desea ejecutar: ')

# Condicionales para responder según la opción elegida
if opcionElegida == '1':
    print('Usted ha elegido la opción', opcionElegida)
elif opcionElegida == '2':
    print('Usted ha elegido la opción', opcionElegida)
elif opcionElegida == '3':
    print('Usted ha elegido la opción', opcionElegida)
elif opcionElegida == '4':
    print('Usted ha elegido la opción', opcionElegida)
else:
    print('Usted ha elegido una opción diferente a las del menú')
```
