## Ejercicio No. 6

1. Variables

    - usuarioGuardado y claveGuardada almacenan las credenciales correctas.

    - usuarioIngresado y claveIngresada capturan lo que escribe el usuario.

2. Validación del usuario

    - Se compara usuarioIngresado con usuarioGuardado.

    - Si no coincide, el programa indica que el usuario no está registrado.

3. Validación de la clave con límite de intentos

    - Se usa un bucle while que permite al usuario ingresar la clave hasta 3 veces.

    - Si la clave coincide con claveGuardada, se concede el acceso y el bucle termina.

    - Si la clave es incorrecta, se incrementa el contador de intentos y se muestra un mensaje.

    - Si se superan los 3 intentos, el programa bloquea el acceso.

4. Control de flujo

    - La combinación de if y while permite validar usuario y clave de manera segura.

    - Se asegura que el programa sea interactivo y limite los intentos de forma controlada.

```python
# Variables con usuario y clave almacenados
usuarioGuardado = "admin"
claveGuardada = "1234"

# Solicitar el nombre de usuario
usuarioIngresado = input("Ingrese su usuario: ")

# Verificar si el usuario es válido
if usuarioIngresado == usuarioGuardado:
    # Inicializar contador de intentos de clave
    intentos = 0
    claveCorrecta = False
    
    # Bucle para ingresar la clave hasta 3 intentos
    while intentos < 3 and not claveCorrecta:
        claveIngresada = input("Ingrese su clave: ")
        if claveIngresada == claveGuardada:
            print("¡Acceso concedido! Bienvenido,", usuarioIngresado)
            claveCorrecta = True
        else:
            intentos += 1
            print(f"Clave incorrecta. Intento {intentos} de 3.")
    
    # Mensaje si se agotaron los intentos
    if not claveCorrecta:
        print("Ha excedido el número de intentos. Acceso denegado.")
else:
    print("Usuario no encontrado. Por favor, regístrese primero.")
    input('Desea registrar un usuario nuevo?')
```
