1. Bucle while

    - Un bucle while permite repetir un bloque de código mientras se cumpla una condición.

    - En este programa, while opcionElegida != '5': mantiene el menú activo hasta que el usuario elija la opción 5 (Salir).

2. Funcionalidad del menú repetitivo

    - Cada vez que el bucle se ejecuta, el menú se imprime y se solicita la elección del usuario.

    - Esto permite que el usuario seleccione diferentes opciones varias veces sin reiniciar el programa.

3. Opción de salida

    - Al elegir 5, la condición del while deja de cumplirse, el bucle termina y el programa muestra un mensaje de despedida.

    - Esto permite que el menú sea interactivo y que el usuario tenga control sobre cuándo finalizar la aplicación.

4. Integración con otros conceptos

    - Se siguen usando tuplas para almacenar las opciones y for para recorrerlas.

    - Las condicionales (if, elif, else) permiten responder según la opción elegida.

```python
# Mensaje de bienvenida
Bienvenida = '\nBienvenidos a la pizzería de Sys On Chip\n'
print(Bienvenida)

# Opciones del menú
opcion1 = '1. Gestión de pedidos'
opcion2 = '2. Gestión de clientes'
opcion3 = '3. Gestión de empleados'
opcion4 = '4. Gestión de inventario'
opcion5 = '5. Salir'

# Agrupamos las opciones en una tupla
menu = (opcion1, opcion2, opcion3, opcion4, opcion5)

# Inicializamos la opción del usuario
opcionElegida = ''

# Bucle principal del menú
while opcionElegida != '5':
    # Mostrar todas las opciones del menú
    print('\nMenú principal\n')
    for elemento in menu:
        print(elemento)
    
    # Solicitar al usuario que elija una opción
    opcionElegida = input('\nQué opción desea ejecutar: ')
    
    # Responder según la opción elegida
    if opcionElegida == '1':
        print('Usted ha elegido la opción', opcion1)
    elif opcionElegida == '2':
        print('Usted ha elegido la opción', opcion2)
    elif opcionElegida == '3':
        print('Usted ha elegido la opción', opcion3)
    elif opcionElegida == '4':
        print('Usted ha elegido la opción', opcion4)
    elif opcionElegida == '5':
        print('Gracias por usar el sistema. ¡Hasta luego!')
    else:
        print('Usted ha elegido una opción diferente a las del menú')
```
