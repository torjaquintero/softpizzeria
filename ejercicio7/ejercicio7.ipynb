1. Funciones en Python

    - Una función es un bloque de código que realiza una tarea específica y puede ser reutilizado.

    - Se define con def nombreFuncion(): y puede recibir parámetros y devolver valores con return.

    - En este ejemplo, registrarUsuario() pide un nombre de usuario y una clave, y devuelve ambos valores para actualizarlos en el programa principal.

2. Flujo del programa

    - Primero se solicita el nombre de usuario.

    - Si coincide con el usuario guardado, se valida la clave con un bucle while que permite hasta 3 intentos.

    - Si no coincide, se llama a la función registrarUsuario() para crear un nuevo usuario y clave.

    - Esto demuestra cómo reutilizar código con funciones y mantener organizado el programa.

3. Interactividad y seguridad

    - Se mantiene la interacción con el usuario y el control de intentos para la contraseña.

    - El programa permite actualizar las credenciales si el usuario no estaba registrado previamente.

```python
# Variables iniciales para almacenar usuario y clave
usuarioGuardado = "admin"
claveGuardada = "1234"

# Función para registrar un nuevo usuario
def registrarUsuario():
    nuevoUsuario = input("Ingrese un nuevo nombre de usuario: ")
    nuevaClave = input("Ingrese una nueva clave: ")
    print(f"Usuario '{nuevoUsuario}' registrado correctamente.")
    return nuevoUsuario, nuevaClave

# Programa principal
usuarioIngresado = input("Ingrese su usuario: ")

if usuarioIngresado == usuarioGuardado:
    intentos = 0
    claveCorrecta = False
    
    while intentos < 3 and not claveCorrecta:
        claveIngresada = input("Ingrese su clave: ")
        if claveIngresada == claveGuardada:
            print("¡Acceso concedido! Bienvenido,", usuarioIngresado)
            claveCorrecta = True
        else:
            intentos += 1
            print(f"Clave incorrecta. Intento {intentos} de 3.")
    
    if not claveCorrecta:
        print("Ha excedido el número de intentos. Acceso denegado.")
else:
    print("Usuario no encontrado. Debe registrarse primero.")
    usuarioGuardado, claveGuardada = registrarUsuario()
```
