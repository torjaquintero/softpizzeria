## Ejemplo No. 2

1. Mensaje de bienvenida: Se almacena en la variable Bienvenida y se imprime usando print(). Los saltos de línea (\n) mejoran la presentación visual en la consola.

2. Menú principal: Se muestran las opciones disponibles para el usuario, cada una guardada en variables (opcion1 y opcion2) y luego impresas.

3. Entrada del usuario: La función input() permite que el usuario escriba un valor desde la consola. Este valor se guarda en la variable opcionElegida.

4. Mostrar la opción seleccionada: Finalmente, se combina un texto fijo con la variable

```python
# Mensaje de bienvenida
Bienvenida = '\nBienvenidos a la pizzería de Sys On Chip\n'
print(Bienvenida)

# Título del menú
print('Menú principal\n')

# Opciones del menú
opcion1 = '1. Gestión de pedidos'
opcion2 = '2. Gestión de clientes'

# Imprimir las opciones
print(opcion1)
print(opcion2)

# Línea en blanco para separar la entrada del usuario
print('\n')

# Solicitar al usuario que elija una opción
opcionElegida = input('Qué opción desea elegir para ejecutar: ')

# Mostrar la opción elegida
print('Usted ha elegido la opción', opcionElegida)
```
