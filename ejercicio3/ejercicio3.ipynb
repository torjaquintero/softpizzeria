## Ejemplo No. 3

1. Menú con varias opciones: Se definen varias opciones en variables (opcion1 a opcion4) y se muestran al usuario usando print().

2. Entrada del usuario: input() permite que el usuario escriba la opción deseada, que se almacena en opcionElegida.

3. Condicionales:

    - if opcionElegida == '1': verifica si el usuario eligió la primera opción.

    - elif opcionElegida == '2': y las siguientes verifican las demás opciones de manera secuencial.

    - else: captura cualquier valor que no coincida con las opciones del menú.

4. Funcionamiento: Python evalúa los condicionales en orden. La primera condición verdadera ejecuta su bloque de código y luego salta el resto. Esto permite que el programa responda de manera diferente según lo que ingrese el usuario.

Con esto, los estudiantes aprenden a tomar decisiones en su programa, lo que es fundamental para hacer aplicaciones interactivas y dinámicas.

```python
# Mensaje de bienvenida
Bienvenida = '\nBienvenidos a la pizzería de Sys On Chip\n'
print(Bienvenida)

# Título del menú
print('Menú principal\n')

# Opciones del menú
opcion1 = '1. Gestión de pedidos'
opcion2 = '2. Gestión de clientes'
opcion3 = '3. Gestión de empleados'
opcion4 = '4. Gestión de inventario'

# Imprimir las opciones
print(opcion1)
print(opcion2)
print(opcion3)
print(opcion4)

# Línea en blanco para separar la entrada del usuario
print('\n')

# Solicitar al usuario que elija una opción
opcionElegida = input('Qué opción desea ejecutar: ')

# Condicionales para responder según la opción elegida
if opcionElegida == '1':
    print('Usted ha elegido la opción', opcion1)
elif opcionElegida == '2':
    print('Usted ha elegido la opción', opcion2)
elif opcionElegida == '3':
    print('Usted ha elegido la opción', opcion3)
elif opcionElegida == '4':
    print('Usted ha elegido la opción', opcion4)
else:
    print('Usted ha elegido una opción diferente a las del menú')
```
