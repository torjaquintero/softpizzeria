Concepto de archivos en Python

1. Archivos

    - Un archivo es un medio para guardar información de manera permanente en el disco.

    - En Python, se puede leer (read) o escribir (write) en archivos usando la función open().

2. CSV (Comma Separated Values)

    - Es un archivo de texto donde cada línea representa un registro y cada campo está separado por comas.

    - Ideal para guardar datos de diccionarios o tablas de usuarios.

3. Módulo csv

    - Python tiene un módulo llamado csv que facilita leer y escribir archivos CSV de manera estructurada.

```python
import csv

# Diccionario ejemplo con un usuario
baseUsuarios = {
    "admin": {
        "nombre_usuario": "admin",
        "clave": 1234,
        "nombre": "Héctor",
        "id": 1,
        "correo": "hector@example.com",
        "celular": "3001234567"
    }
}

# Guardar el diccionario en un archivo CSV
with open("baseUsuarios.csv", mode="w", newline="") as archivo:
    campos = ["nombre_usuario", "clave", "nombre", "id", "correo", "celular"]
    escritor = csv.DictWriter(archivo, fieldnames=campos)
    
    escritor.writeheader()  # escribir los nombres de las columnas
    for usuario in baseUsuarios.values():
        escritor.writerow(usuario)  # escribir cada usuario como una fila

print("Diccionario guardado en baseUsuarios.csv correctamente.")
```

Qué introduce este programa

1. Diccionarios → para manejar toda la información de los usuarios.

2. Archivos CSV → para guardar y cargar los usuarios (baseUsuarios.csv).

3. Login y registro → con clave de 4 dígitos y manejo de errores (try / except).

4. Bucle while y for → para el menú interactivo y validación de entradas.

5. Funciones → para organizar el código: registrarUsuario(), login(), menuPrincipal(), cargarUsuarios(), guardarUsuarios().

```python
import csv

# -----------------------------
# Función para cargar usuarios desde archivo CSV
# -----------------------------
def cargarUsuarios():
    usuarios = {}
    try:
        with open("baseUsuarios.csv", mode="r", newline="") as archivo:
            lector = csv.DictReader(archivo)
            for fila in lector:
                # Convertir clave e id a enteros
                fila["clave"] = int(fila["clave"])
                fila["id"] = int(fila["id"])
                usuarios[fila["nombre_usuario"]] = fila
        print("Usuarios cargados desde baseUsuarios.csv.")
    except FileNotFoundError:
        print("Archivo baseUsuarios.csv no encontrado. Se creará al registrar el primer usuario.")
    return usuarios

# -----------------------------
# Función para guardar usuarios en archivo CSV
# -----------------------------
def guardarUsuarios(usuarios):
    campos = ["nombre_usuario", "clave", "nombre", "id", "correo", "celular"]
    with open("baseUsuarios.csv", mode="w", newline="") as archivo:
        escritor = csv.DictWriter(archivo, fieldnames=campos)
        escritor.writeheader()
        for usuario in usuarios.values():
            escritor.writerow(usuario)
    print("Usuarios guardados en baseUsuarios.csv correctamente.")

# -----------------------------
# Función para registrar un nuevo usuario con doble confirmación de clave
# -----------------------------
def registrarUsuario():
    print("\n--- Registro de nuevo usuario ---")
    nombre_usuario = input("Ingrese un nombre de usuario: ")

    # Clave: solo números enteros de 4 dígitos y se confirma dos veces
    while True:
        clave1 = input("Ingrese una clave (4 dígitos): ")
        clave2 = input("Confirme su clave: ")
        try:
            clave1 = int(clave1)
            clave2 = int(clave2)
            if clave1 != clave2:
                print("Error: las claves no coinciden. Intente nuevamente.")
            elif not (1000 <= clave1 <= 9999):
                print("Error: la clave debe tener exactamente 4 dígitos.")
            else:
                clave = clave1  # clave válida y confirmada
                break
        except ValueError:
            print("Error: debe ingresar un número entero.")

    nombre = input("Ingrese su nombre completo: ")
    
    while True:
        try:
            numero_id = int(input("Ingrese su número de identificación: "))
            break
        except ValueError:
            print("Error: debe ingresar un número entero.")
    
    correo = input("Ingrese su correo electrónico: ")
    celular = input("Ingrese su número de celular: ")
    
    usuario = {
        "nombre_usuario": nombre_usuario,
        "clave": clave,
        "nombre": nombre,
        "id": numero_id,
        "correo": correo,
        "celular": celular
    }
    
    print(f"\nUsuario '{nombre_usuario}' registrado correctamente.\n")
    return usuario

# -----------------------------
# Función de login
# -----------------------------
def login(usuarios):
    usuarioIngresado = input("Ingrese su usuario: ")

    if usuarioIngresado in usuarios:
        intentos = 0
        claveCorrecta = False
        while intentos < 3 and not claveCorrecta:
            try:
                claveIngresada = int(input("Ingrese su clave (4 dígitos): "))
                if 1000 <= claveIngresada <= 9999:
                    if claveIngresada == usuarios[usuarioIngresado]["clave"]:
                        print(f"¡Acceso concedido! Bienvenido {usuarios[usuarioIngresado]['nombre']}\n")
                        claveCorrecta = True
                    else:
                        intentos += 1
                        print(f"Clave incorrecta. Intento {intentos} de 3.")
                else:
                    print("Error: la clave debe tener exactamente 4 dígitos.")
            except ValueError:
                print("Error: debe ingresar un número entero.")
        if not claveCorrecta:
            print("Ha excedido el número de intentos. Acceso denegado.\n")
            return None
        return usuarioIngresado
    else:
        while True:
            respuesta = input("Usuario no encontrado. ¿Desea registrarse? (s/n): ").lower()
            if respuesta == 's':
                nuevoUsuario = registrarUsuario()
                usuarios[nuevoUsuario["nombre_usuario"]] = nuevoUsuario
                guardarUsuarios(usuarios)
                return nuevoUsuario["nombre_usuario"]
            elif respuesta == 'n':
                print("No se registró ningún usuario. Programa finalizado.\n")
                return None
            else:
                print("Error: solo puede ingresar 's' o 'n'.")

# -----------------------------
# Función para mostrar menú principal
# -----------------------------
def menuPrincipal():
    menu = (
        "1. Gestión de pedidos",
        "2. Gestión de clientes",
        "3. Gestión de empleados",
        "4. Gestión de inventario",
        "5. Salir"
    )
    
    while True:
        print("\n--- Menú Principal ---")
        for opcion in menu:
            print(opcion)
        opcionElegida = input("\nQué opción desea ejecutar: ")
        
        if opcionElegida == '1':
            print("Usted ha elegido: Gestión de pedidos")
        elif opcionElegida == '2':
            print("Usted ha elegido: Gestión de clientes")
        elif opcionElegida == '3':
            print("Usted ha elegido: Gestión de empleados")
        elif opcionElegida == '4':
            print("Usted ha elegido: Gestión de inventario")
        elif opcionElegida == '5':
            print("Saliendo del sistema...")
            break
        else:
            print("Opción no válida. Por favor ingrese un número del 1 al 5.")

# -----------------------------
# Programa principal
# -----------------------------
usuarios = cargarUsuarios()  # cargar usuarios desde archivo

usuarioActual = login(usuarios)  # realizar login o registro

if usuarioActual:
    menuPrincipal()  # mostrar menú principal si login fue exitoso
```
