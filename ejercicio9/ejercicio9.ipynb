1. Concepto de diccionario

    - Un diccionario es una colección de elementos donde cada clave tiene un valor asociado.

    - Se define con {} y se accede mediante la clave.

    - Ejemplo simple:

```python
usuario = {
    "nombre_usuario": "admin",
    "clave": 1234,
    "nombre": "Héctor",
    "id": 1,
    "correo": "hector@example.com",
    "celular": "3001234567"
}

print(usuario["nombre"])  # salida: Héctor
```

Qué hace este código:

1. Diccionario baseUsuarios

    - Almacena todos los usuarios del sistema.

    - La clave del diccionario principal es nombre_usuario.

    - El valor es otro diccionario con todos los datos del usuario.

2. Registro de usuario

    - La función registrarUsuario() solicita nombre de usuario, clave de 4 dígitos, nombre completo, ID, correo y celular.

    - Devuelve un diccionario con todos los datos del usuario.

3. Primer usuario

    - Si baseUsuarios está vacío, el programa obliga a crear el primer usuario.

    - Este usuario se guarda en el diccionario general, usando su nombre_usuario como clave.

4. Recorrer diccionario

    - for clave, valor in baseUsuarios.items(): permite imprimir todos los usuarios registrados y sus datos.

```python
# -----------------------------
# Diccionario base de usuarios
# -----------------------------
baseUsuarios = {}  # inicialmente vacío, se llenará con el primer usuario

# -----------------------------
# Función para registrar un nuevo usuario
# -----------------------------
def registrarUsuario():
    print("\n--- Registro de nuevo usuario ---")
    
    nombre_usuario = input("Ingrese un nombre de usuario: ")
    
    # Validar clave de 4 dígitos
    while True:
        clave = input("Ingrese una clave (4 dígitos): ")
        try:
            clave = int(clave)
            if 1000 <= clave <= 9999:
                break
            else:
                print("Error: la clave debe tener exactamente 4 dígitos.")
        except ValueError:
            print("Error: debe ingresar un número entero.")
    
    # Datos adicionales
    nombre = input("Ingrese su nombre completo: ")
    while True:
        try:
            numero_id = int(input("Ingrese su número de identificación: "))
            break
        except ValueError:
            print("Error: debe ingresar un número entero.")
    
    correo = input("Ingrese su correo electrónico: ")
    celular = input("Ingrese su número de celular: ")
    
    # Crear diccionario del usuario
    usuario = {
        "nombre_usuario": nombre_usuario,
        "clave": clave,
        "nombre": nombre,
        "id": numero_id,
        "correo": correo,
        "celular": celular
    }
    
    print(f"\nUsuario '{nombre_usuario}' registrado correctamente.\n")
    return usuario

# -----------------------------
# Programa principal
# -----------------------------
print("Bienvenido al sistema de usuarios.")

# Verificar si baseUsuarios está vacío
if not baseUsuarios:
    print("No hay usuarios registrados. Debe crear el primer usuario.")
    primerUsuario = registrarUsuario()
    # Usamos el nombre de usuario como clave en el diccionario general
    baseUsuarios[primerUsuario["nombre_usuario"]] = primerUsuario

# Mostrar la base de usuarios
print("\nBase de usuarios actual:")
for clave, valor in baseUsuarios.items():
    print(f"{clave} => {valor}")
```
