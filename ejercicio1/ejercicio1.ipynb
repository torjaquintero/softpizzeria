Bienvenidos al curso de introducción a python de Sys On Chip

En este programa usamos la función print() para mostrar mensajes en la consola. Observa que dentro de las cadenas de texto se incluye \n, que es un carácter especial llamado salto de línea. Este carácter le indica a Python que debe comenzar una nueva línea al mostrar el texto.

La primera línea imprime un mensaje de bienvenida, con un salto de línea antes y después, lo que hace que el mensaje aparezca centrado visualmente y separado de otros mensajes.

La segunda línea presenta la actividad que realizaremos en el curso: construir un software para una pizzería.

El uso de print() con saltos de línea es muy útil para organizar la información que se muestra al usuario y hacer que la salida del programa sea más clara y agradable a la vista.

```python
# Imprimimos un mensaje de bienvenida con un salto de línea antes y después
print('\nHola, bienvenidos al curso de Python\n')

# Imprimimos otro mensaje indicando la primera actividad
print('Vamos a construir un software para una pizzería\n')
```

En este ejemplo introducimos las variables, que son espacios de memoria donde podemos almacenar información para usarla después.

En la línea mensaje = 'Bienvenidos a la pizzería Sys On Chip', estamos creando una variable llamada mensaje y asignándole un texto.

Luego, usamos print(mensaje) para mostrar el contenido de la variable en la consola. Python imprime exactamente lo que está almacenado en mensaje.

El uso de variables permite que nuestros programas sean más flexibles y reutilizables. Por ejemplo, si queremos cambiar el mensaje de bienvenida, solo necesitamos modificar el contenido de la variable sin tocar la función print(). Esto es mucho más eficiente que escribir directamente el texto cada vez.

```python
# Guardamos un mensaje en una variable
mensaje = 'Bienvenidos a la pizzería Sys On Chip'

# Imprimimos el contenido de la variable
print(mensaje)
```
